```python
import pandas as pd

df = pd.read_csv("economy_total_current_0820.csv")
display(df)
```

```python

##CHAT_GPT로 짰습니다
def price_to_int(price_str):
    # Remove '편도', ',', '원~', and convert to integer
    return int(price_str.replace('편도', '').replace(',', '').replace('원~', '').strip())

# Apply the function to the 'price' column
df['price'] = df['price'].apply(price_to_int)

# Function to convert time string to minutes
def time_to_min(time_str):
    # Split the string into hours and minutes
    hours, minutes = map(int, time_str.replace('시간', '').replace('분', '').split())
    # Convert to minutes
    return hours * 60 + minutes


#직항이면 0, 경유1이면 1, 경유 2 이상이면 2 반환하는 함수
def dir_to_int(dir_str):
    if dir_str =='직항':
        return 0
    elif dir_str =='경유 1':
        return 1
    else:
        return 2



# Apply the function to the 'time_dir' column
df['time_dir'] = df['time_dir'].apply(time_to_min)
# 코드에 잘못된 부분 수정(time_taken과 time_dir 이 잘못되어 있음)
df['time_dir'],df['time_taken'] = df['time_taken'],df['time_dir']
df['time_dir'] = df['time_dir'].apply(dir_to_int)
# Display the first few rows of the DataFrame
df.head()
```
