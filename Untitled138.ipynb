```python
import pandas as pd



df = pd.read_csv("/content/data_merge_final.csv",encoding='cp949')
```

```python
display(df)


def price_to_int(price_str):

    return int(price_str.replace('편도', '').replace(',', '').replace('원~', '').strip())


df['가격'] = df['가격'].apply(price_to_int)


def time_to_min(time_str):

    hours, minutes = map(int, time_str.replace('시간', '').replace('분', '').split())

    return hours * 60 + minutes



def dir_to_int(dir_str):
    if dir_str =='직항':
        return 0
    elif dir_str =='경유 1':
        return 1
    else:
        return 2

df['총 소요 시간'] = df['총 소요 시간'].apply(time_to_min)
df['항공편'] = df['항공편'].apply(dir_to_int)
display(df)
```
